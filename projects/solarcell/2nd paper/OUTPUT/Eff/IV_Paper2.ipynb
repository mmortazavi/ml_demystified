```python
import numpy as np
import pandas as pd
```

#Eff calculation

#sample1

```python
s1 = pd.read_excel('IV.xlsx', sheet_name='Sample #1', header=None)
s1.dropna(axis='columns',inplace=True)
s1
```

```python
#Power
bare = s1.iloc[:,2:3]
drop1 = s1.iloc[:, 5:6]
drop2 = s1.iloc[:,8:9]
drop3 = s1.iloc[:,11:12]
```

```python
max_bare = np.max(bare)
max_drop1 = np.max(drop1)
max_drop2 = np.max(drop2)
max_drop3 = np.max(drop3)


#  DataFrame
s11 = {
    'P-max/10': [max_bare/10, max_drop1/10, max_drop2/10, max_drop3/10],
    'P-max/12': [max_bare/12, max_drop1/12, max_drop2/12, max_drop3/12]
}
Sample1 = pd.DataFrame(s11, index=['bare', 'drop1', 'drop2', 'drop3',])


Sample1['Δeff(A=10)'] = Sample1['P-max/10'] - Sample1.loc['bare', 'P-max/10']
Sample1['Δeff(A=12)'] = Sample1['P-max/12'] - Sample1.loc['bare', 'P-max/12']
Sample1
```
