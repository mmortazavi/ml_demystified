```python
#theta

for i in range(1, 7):  # 6 peaks
    theta_col = f'2-theta {i}'

    # Calculate theta (2Theta / 2)
    df[f'Theta_{i}'] = df[theta_col] / 2
```

```python
def cos_degrees(angle_degrees):
    angle_radians = math.radians(angle_degrees)
    return math.cos(angle_radians)
```

```python
cos_theta_cols = df[['Theta_1', 'Theta_2', 'Theta_3', 'Theta_4', 'Theta_5', 'Theta_6']].applymap(cos_degrees)

df = pd.concat([df, cos_theta_cols.add_prefix('Cos_')], axis=1)
```

```python
df
```

```python
# Function to calculate Sherer parameter

def calculate_sherer(fwhm, cos_theta, lamda, k):
    return k * lamda / (fwhm * cos_theta)

# Function to calculate dislocation density
def calculate_dislocation_density(sherer):
    return 1 / sherer**2


# Constants
lamda = 1.54  # Wavelength in Angstrom
k = 0.089  # Constant


# Calculate Sherer parameter for each peak in each sample
for i in range(1, 7):  # Assuming 3 peaks for illustration
    fwhm_col = f'FWHM {i}'
    cos_theta_col = f'Cos_Theta_{i}'
    sherer_col = f'Sherer_{i}'
    dislocation_density_col = f'Dislocation_Density_{i}'

    df[sherer_col] = calculate_sherer(df[fwhm_col], df[cos_theta_col], lamda, k)
    df[dislocation_density_col] = calculate_dislocation_density(df[sherer_col])
```
