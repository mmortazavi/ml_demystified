```python
import numpy as np
import pandas as pd
```

#funcs

```python
def IV(df):

    df.iloc[:, 1:2] = df.iloc[:, 1:2].apply(lambda x: x * -1)

    df = df[df.iloc[:, 0:3] >= 0]
    df = df.dropna(axis=0, how='any')

    v = df.iloc[:, 0:1].to_numpy().flatten()
    i = df.iloc[:, 1:2].to_numpy().flatten()
    p = df.iloc[:, 2:3].to_numpy().flatten()

    isc = i[0]  # Short circuit current
    voc = v[-1]  # Open circuit voltage

    p = v * i
    idx = np.argmax(p)
    p_max = np.max(p)

    ff = (v[idx] * i[idx]) / (v[-1] * i[0])

    result = {
        'voc': voc,
        'isc': isc,
        'P_max': p_max,
        #'V_mp': v[idx],
        #'I_mp': i[idx],
        'Eff A=12': p_max / 12,
        'Eff A=10': p_max / 10,
        'ff*100': ff * 100,
        'ff*1.2*100': ff * 1.2 * 100
    }


    result_df = pd.DataFrame([result])

    return result_df
```

```python
def process_samples(s):
    # Slice
    bare = s.iloc[:,:3]
    drop1 = s.iloc[:, 3:6]
    drop2 = s.iloc[:,6:9]
    drop3 = s.iloc[:,9:12]
    drop4 = s.iloc[:, 12:15]
    drop5 = s.iloc[:, 15:18]
    drop6 = s.iloc[:,18:21]

    # Calculate IV params for each gp
    d0 = IV(bare)
    d1 = IV(drop1)
    d2 = IV(drop2)
    d3 = IV(drop3)
    d4 = IV(drop4)
    d5 = IV(drop5)
    d6 = IV(drop6)

    #output
    sample = pd.concat([d0, d1, d2, d3, d4, d5, d6], keys=['bare', 'drop1', 'drop2', 'drop3', 'drop4', 'drop5', 'drop6'])

    sample = sample.round(2)

    return sample

```
