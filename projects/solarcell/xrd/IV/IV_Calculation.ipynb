```python
import numpy as np
import pandas as pd
```

#funcs

```python
def IV(df):


    # Negate the values in the second (current) and third (power) columns
    df.iloc[:, 1:3] *= -1

    # keep only positive values in the current column
    df = df[df.iloc[:, 1] >= 0]

    # Drop missing values ( just in case !)
    df = df.dropna()

    # Calculate parameters
    v, i, p = df.iloc[:, 0].values, df.iloc[:, 1].values, df.iloc[:, 2].values
    isc, voc = i[0], v[-1]
    p = v * i
    idx, p_max = np.argmax(p), np.max(p)
    ff = (v[idx] * i[idx]) / (voc * isc)

    #  calculated values
    result = {
        'voc': voc,
        'isc': isc,
        'P_max': p_max,
        'V_mp': v[idx],
        'I_mp': i[idx],
        'Eff A=12': p_max / 12,
        'Eff A=10': p_max / 10,
        'ff*100': ff * 100,
        'ff*1.2*100': ff * 1.2 * 100
    }

    # Output dataFrame
    result_df = pd.DataFrame([result])

    return result_df
```

```python
def process_samples(s):

    # slice each sample to subsets based on drop condition
    bare = s.iloc[:,:3]
    drop1 = s.iloc[:, 3:6]
    drop2 = s.iloc[:,6:9]
    drop3 = s.iloc[:,9:12]
    drop4 = s.iloc[:, 12:15]
    drop5 = s.iloc[:, 15:18]
    drop6 = s.iloc[:,18:21]

    # Calculate IV params for each subset
    sample_data = []
    if not bare.empty:
        sample_data.append(('bare', IV(bare)))
    if not drop1.empty:
        sample_data.append(('drop1', IV(drop1)))
    if not drop2.empty:
        sample_data.append(('drop2', IV(drop2)))
    if not drop3.empty:
        sample_data.append(('drop3', IV(drop3)))
    if not drop4.empty:
        sample_data.append(('drop4', IV(drop4)))
    if not drop5.empty:
        sample_data.append(('drop5', IV(drop5)))
    if not drop6.empty:
        sample_data.append(('drop6', IV(drop6)))

    # Output
    sample = pd.concat([data[1] for data in sample_data], keys=[data[0] for data in sample_data])

    sample = sample.round(2)

    return sample
```
