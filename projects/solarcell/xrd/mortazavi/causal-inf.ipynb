```python
data = data[['DROP', 
             'relative_intensity 1','relative_intensity 2', 'relative_intensity 3', 'relative_intensity 4', 'relative_intensity 5', 'relative_intensity 6',       
                'FF']]
```

```python


# Convert your DataFrame to a NumPy array
data_np = data.to_numpy()

# Apply the PC algorithm
# The PC algorithm requires a significance level (alpha). Typically, 0.05 is used.
alpha = 0.05
cg = pc(data_np, alpha)

# The result is a causal graph object
# Plot the resulting causal graph
# plot_graph(cg, labels=data.columns)
# visualization using pydot
# or save the graph
# from causallearn.utils.GraphUtils import GraphUtils

# pyd = GraphUtils.to_pydot(cg.G)
# pyd.write_png('simple_test.png')
```
