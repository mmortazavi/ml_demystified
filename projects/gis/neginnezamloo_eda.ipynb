```python
import pandas as pd
import plotly.express as px
```

```python
preprocessed=pd.read_excel(r'C:\Users\Asus\Desktop\preprocessed_data_ngnz.xlsx') 
preprocessed.drop(columns=[ 'Unnamed: 0'] , inplace=True)
```

```python
preprocessed.info()
```

```python
preprocessed.drop(preprocessed.loc[preprocessed['postal_code']==1111111111].index,inplace=True)
```

```python
# Lines on Mapbox
fig = px.line_mapbox(preprocessed, lat='φ(d)_by_postal_code',lon='λ(d)_by_postal_code',
                     hover_name="shahr_e_code_posti", zoom=2, height=800)

fig.update_layout(mapbox_style="stamen-terrain", mapbox_zoom=4, mapbox_center_lat = 41)


fig.show()
```

```python
# Mapbox Density Heatmap
fig = px.density_mapbox(preprocessed, lat='φ(d)_by_postal_code',lon='λ(d)_by_postal_code', 
                         
                        zoom=3,hover_name="shahr_e_code_posti",height=800,radius=8,
                         
                        mapbox_style="stamen-terrain")
fig.show()
```
